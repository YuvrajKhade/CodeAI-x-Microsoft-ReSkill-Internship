```python
import pandas as pd
```

```python
data=pd.read_csv('/content/drive/MyDrive/books-dataset-2.csv')
```

```python
data.head()
```

```python
def search_book(keyword):
    keyword = keyword.lower()
    result = data[
        data['title'].str.lower().str.contains(keyword) |
        data['authors'].str.lower().str.contains(keyword) |
        data['publisher'].str.lower().str.contains(keyword)
    ]
    return result

def display_book(data):
    if data.empty:
        print("No book found")
    else:
        for index, row in data.iterrows():
            print(f"Title: {row['title']}")
            print(f"Authors: {row['authors']}")
            print(f"Average rating: {row['average_rating']}")
            print(f"ISBN: {row['isbn']}")
            print(f"ISBN13: {row['isbn13']}")
            print(f"Language: {row['language_code']}")
            if 'num_pages' in data.columns:
                print(f"Number of pages: {row['num_pages']}")
            print(f"Rating count: {row['ratings_count']}")
            print(f"Text reviews count: {row['text_reviews_count']}")
            print(f"Publication date: {row['publication_date']}")
            print(f"Publisher: {row['publisher']}")
            print('-' * 40)

def main():
    print("Welcome to book recommendation system")
    keyword = input("Enter a keyword: ")
    result = search_book(keyword)
    display_book(result)

if __name__ == "__main__":
    main()
```
